## Home vs Neutral Venue Factor

```python
def win_fun(x):
    if x['home_score']>x['away_score']:
        return x['home_team']
    elif x['home_score']==x['away_score']:
        return 'Draw'
    else:
        return x['away_team']
```

```python
data['winner'] = data.apply(lambda x: win_fun(x), axis=1)
```

```python
england_home_wins = data[np.logical_and(np.logical_and(data['home_team']=='England', data['winner']=='England'), data['country']=='England')].shape[0]
england_away_wins =  data[np.logical_and(np.logical_and(data['away_team']=='England', data['winner']=='England'), data['country']!='England')].shape[0]
print(england_home_wins)
england_away_wins
```

```python
home_wins, away_wins = [], []
home_games, away_games = [], []

for i in country_df['country']:
    h_win = data[np.logical_and(np.logical_and(data['home_team']==i, data['winner']==i), data['country']==i)].shape[0]
    a_win = data[np.logical_and(np.logical_and(data['away_team']==i, data['winner']==i), data['country']!=i)].shape[0]
    h_games = data[np.logical_and(data['home_team']==i, data['country']==i)].shape[0]
    a_games = data[np.logical_and(data['away_team']==i, data['country']!=i)].shape[0]
    home_wins.append(h_win), away_wins.append(a_win), home_games.append(h_games), away_games.append(a_games)
```

```python
country_df['home_wins'] = home_wins
country_df['away_wins'] = away_wins
country_df['home_games'] = home_games
country_df['away_games'] = away_games
```

```python
country_df
```

```python
country_df['h_wins_%'] = country_df.apply(lambda x: (x['home_wins']/max(1,x['home_games']))*100, axis=1)
country_df['a_wins_%'] = country_df.apply(lambda x: (x['away_wins']/max(1,x['away_games']))*100, axis=1)
country_df
```
